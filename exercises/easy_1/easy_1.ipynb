##  1 - Banner Class

### Problem Description



```ruby
# Given Code

class Banner
  def initialize(message)
  end

  def to_s
    [horizontal_rule, empty_line, message_line, empty_line, horizontal_rule].join("\n")
  end

  private

  def horizontal_rule
  end

  def empty_line
  end

  def message_line
    "| #{@message} |"
  end
end
```

```ruby
# My Solution

class Banner
  def initialize(message)
    self.message = message
  end

  def to_s
    [horizontal_rule, empty_line, message_line, empty_line, horizontal_rule].join("\n")
  end

  private

  attr_accessor :message
  
  def horizontal_rule
    "+-#{'-' * message.size}-+"
  end

  def empty_line
    "| #{' ' * message.size} |"
  end

  def message_line
    "| #{message} |"
  end
end
```

```ruby
puts Banner.new('Hamachi: Neko Ichiban Desu')
```

```ruby
# Test Cases

banner = Banner.new('To boldly go where no one has gone before.')
puts banner
# +--------------------------------------------+
# |                                            |
# | To boldly go where no one has gone before. |
# |                                            |
# +--------------------------------------------+
```

```ruby
# Test Cases

banner = Banner.new('')
puts banner
# +--+
# |  |
# |  |
# |  |
# +--+
```

```ruby
# LS Solution

class Banner
  def initialize(message)
    @message = message
  end

  def to_s
    [horizontal_rule, empty_line, message_line, empty_line, horizontal_rule].join("\n")
  end

  private

  def empty_line
    "| #{' ' * (@message.size)} |"
  end

  def horizontal_rule
    "+-#{'-' * (@message.size)}-+"
  end

  def message_line
    "| #{@message} |"
  end
end
```

### Discussion



```ruby
# Further Exploration

class Banner
  def initialize(message, text_width: message.size)
    self.message = message
    self.text_width = text_width
  end

  def to_s
    [horizontal_rule, empty_line, message_line, empty_line, horizontal_rule].join("\n")
  end

  private

  attr_accessor :message, :text_width
  
  def horizontal_rule
    "+-#{'-' * text_width}-+"
  end

  def empty_line
    "| #{' ' * text_width} |"
  end

  def message_line
    "| #{message.center(text_width)} |"
  end
end
```

```ruby
puts Banner.new('Hamachi', text_width: 20)
```

##  5 - Fix the Program - Persons

### Problem Description



```ruby
# Given Code

class Person
  def initialize(first_name, last_name)
    @first_name = first_name.capitalize
    @last_name = last_name.capitalize
  end

  def to_s
    "#{@first_name} #{@last_name}"
  end
end

person = Person.new('john', 'doe')
puts person

person.first_name = 'jane'
person.last_name = 'smith'
puts person
```

```ruby
# Test Cases

# John Doe
# Jane Smith
```

```ruby
# My Solution

class Person
  def initialize(first_name, last_name)
    self.first_name = first_name
    self.last_name = last_name
  end

  def first_name=(first_name)
    @first_name = first_name.capitalize
  end

  def last_name=(last_name)
    @last_name = last_name.capitalize
  end

  def to_s
    "#{@first_name} #{@last_name}"
  end
end

person = Person.new('john', 'doe')
puts person

person.first_name = 'jane'
person.last_name = 'smith'
puts person
```

```ruby
# LS Solution

def first_name= (value)
  @first_name = value.capitalize
end

def last_name= (value)
  @last_name = value.capitalize
end
```
